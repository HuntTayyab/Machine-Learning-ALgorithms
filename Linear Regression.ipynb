```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 

class style:
   BOLD = '\033[1m'
   END = '\033[0m'
    
  
def cf_finder(A, B): 
    
    count = np.size(A);    print("element pair count",count) #number of elements
    MeanA = np.mean(A) ; MeanB=np.mean(B)  
    print("mean of As->", MeanA)
    print("mean of Bs->", MeanB)
    
    AB=np.sum((A-MeanA)*(B-MeanB));    
  
    AA=np.sum((A-MeanA)**2)
    B1 = AB / AA               #co-variance
    #print("B1-",B1)
    B0 = MeanB - B1*MeanA 
    return(B0, B1) 
  
def main(): 
     
    # data = pd.read_csv("filename.csv")  # to read data from CSV
    print(style.BOLD)
    
    # X-axis-A 
    A = np.array([1.1, 1.3, 1.5, 2 ,2.2, 2.9 ,3 ,3.2 ,3.2 ,3.7, 3.9,4 ,4 ,4.1 ,4.5 ,4.9 ,5.1 ,5.3
                   ,5.9 ,6 ,6.8 ,7.1 ,7.9 ,8.2 ,8.7 ,9 ,9.5  ,9.6 ,10.3 ,10.5])
    #Y-axis-B    
    B = np.array([39343 ,46205 ,37731 ,43525 ,39891 ,56642,60150, 54445 ,64445 ,57189 ,63218
                    ,55794 ,56957 ,57081 ,61111 ,67938 ,66029 ,83088 ,81363 ,93940 ,91738 ,98273 ,101302
                    ,113812 ,109431 ,105582 ,116969 ,112635 ,122391 ,121872]) 
    
    # estimating coefficients 
    b = cf_finder(A, B) 
    print("coffeciant B0->",b[0]) 
    print("coffeciant B1->",b[1]) 
    
    plt.scatter(A, B, color = "r")   #to plot scatter graph
    
    y_pred = b[0] + b[1]*A   ## predicted response
    
    #mulit_y_pred=b[0] + b[1]*math.log2(A)
    #print("multi-",mulit_y_pred)
    
    plt.plot(A, y_pred, color = "g", marker = "o")   # to plot mean line
    
    plt.xlabel('A')  # to plot A on X-axis
    plt.ylabel('B')   #to plot B on Y 
    plt.show()
    
    print(style.BOLD)
    #to find error
    yy=b[0] + b[1]*A
    LSE=sum((B-yy)**2)        #least square error
    print("Least Square Error->",LSE)
    
    MAE=sum(abs(B-yy))        #mean absolute error
    print("Mean Absolute Error->",MAE)
    #data=[arr1, arr2]
    
    data1={'Predicate[X]':A,'Response[Y]':B, 'y-cap':yy, 'Error-Diff':B-yy}
    df = pd.DataFrame(data1);
    display(df)
   
if __name__ == "__main__": 
    main() 
```
